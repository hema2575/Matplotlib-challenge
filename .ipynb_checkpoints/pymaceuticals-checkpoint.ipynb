```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
import numpy as np
from scipy.stats import linregress
```

```python
# Study data files
mouse_metadata_path = "Pymaceuticals_data_Mouse_metadata.csv"
study_results_path = "Pymaceuticals_data_Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
pd.set_option('display.max_rows', 2000)
```

```python
print("Here is the data about each mouse :") 
mouse_metadata
```

```python
print("Here are the study results on the mouse :") 
study_results
```

```python
# Combine the data into a single dataset
# Display the data table for preview
mousedata_and_results = pd.merge(mouse_metadata, study_results, how ='outer', on = 'Mouse ID')
pd.set_option('display.max_rows', 2000)
print("Here is the combined dataset :") 
mousedata_and_results
```

```python
#Before beginning the analysis, check the data for any mouse ID with duplicate time points and remove
#any data associated with that mouse ID
#Observations and Insights
```

```python
# Checking the number of mice.
number_of_mice = mousedata_and_results['Mouse ID'].nunique()

print(f" The number of unique Mice are : {number_of_mice}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.(returning dataframe) 
duplicateIDs = mousedata_and_results.loc[mousedata_and_results.duplicated(subset = ['Mouse ID', 'Timepoint']),['Mouse ID','Timepoint']]
duplicateIDs
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. (returning series)
duplicateIDs1 = mousedata_and_results.loc[mousedata_and_results.duplicated(subset = ['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicateIDs1
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows = mousedata_and_results.loc[mousedata_and_results['Mouse ID'] == 'g989']
print(f"Here are all the duplicate rows for mouse id: {duplicateIDs1}") 
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mousestudy_cleaned = mousedata_and_results.loc[mousedata_and_results['Mouse ID'] != 'g989']
print(f"Here is the cleaned dataframe") 
mousestudy_cleaned
```

```python
 # Checking the number of mice in the clean DataFrame.
mousestudy_cleaned['Mouse ID'].nunique()
print(f"The number of unique mice in the cleaned dataframe is :{mousestudy_cleaned['Mouse ID'].nunique()}")
```

```python
#Summary Statistics
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mousestudy_cleaned.groupby('Drug Regimen')[["Tumor Volume (mm3)"]].mean() #--- this produces a dataframe
mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean() #--- this produces a series. We will use series and build the DF.
```

```python
# Assemble the resulting series into a single summary dataframe.
summary1_df = pd.DataFrame({'Mean': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean(), 
                           'Median': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].median(), 
                           'Variance': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].var(),
                           'Std Deviation': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].std(),
                           'SEM': mousestudy_cleaned.groupby('Drug Regimen')["Tumor Volume (mm3)"].sem()
                           })
summary1_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
summary2_df = mousestudy_cleaned.groupby('Drug Regimen').agg({"Tumor Volume (mm3)": ["mean", "median","var", "std","sem"]})
summary3_df = summary2_df.rename(columns={"mean": "Mean", "median": "Median", "var": "Variance", "std": "Std Deviation", "sem": "SEM"})
summary3_df
```

```python
# Quartiles, Outliers and Boxplots
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# 'Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'
# Start by getting the last (greatest) timepoint for each mouse

grtstTimepoint1 = study_results.groupby('Mouse ID').Timepoint.agg(['max'])
grtstTimepoint = grtstTimepoint1.rename(columns = {'max':'Timepoint'})
grtstTimepoint

#maxTimepointforeachmouse = mousestudy_cleaned[mousestudy_cleaned['Timepoint'] == mousestudy_cleaned['Timepoint'].max()] #Thiswill return the entire row with max value 
#maxforselecteddrugs= maxTimepointforeachmouse[maxTimepointforeachmouse['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

#maxTimepointforeachmouse
#maxforselecteddrugs
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint. This
#results in all the mice with all drugregimens and only the greatest time point. 
maxtpTumorVol = pd.merge(grtstTimepoint, mousedata_and_results, on=['Mouse ID','Timepoint'], how = 'left')
maxtpTumorVol
#  Put treatments into a list for for loop (and later for plot labels)
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# This is the dataframe of mice with their tumorvolume at the greatest time point under the 4 selected drug regimen.
maxforselecteddrugs= maxtpTumorVol[maxtpTumorVol['Drug Regimen'].isin(drug_list)]
maxforselecteddrugs

# Creating individual dataframes for each drug regimen.
maxforCapomulin_df = maxforselecteddrugs[maxforselecteddrugs["Drug Regimen"]=='Capomulin']
maxforCapomulin_df

maxforRamicane_df = maxforselecteddrugs[maxforselecteddrugs["Drug Regimen"]=='Ramicane']
maxforRamicane_df

maxforInfubinol_df = maxforselecteddrugs[maxforselecteddrugs["Drug Regimen"]=='Infubinol']
maxforInfubinol_df

maxforCeftamin_df = maxforselecteddrugs[maxforselecteddrugs["Drug Regimen"]=='Ceftamin']
maxforCeftamin_df

# Extracting the series from each of the Dataframes for the boxplot 
Capomulin_tumorVol = maxforCapomulin_df['Tumor Volume (mm3)']
Capomulin_tumorVol

Ramicane_tumorVol = maxforRamicane_df['Tumor Volume (mm3)']
Ramicane_tumorVol

Infubinol_tumorVol = maxforInfubinol_df['Tumor Volume (mm3)']
Infubinol_tumorVol

Ceftamin_tumorVol = maxforCeftamin_df['Tumor Volume (mm3)']
Ceftamin_tumorVol

# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Capomulin quartiles and IQR
cap_quartiles = Capomulin_tumorVol.quantile([.25,.5,.75])
cap_lowerq = cap_quartiles[0.25]
cap_median = cap_quartiles[0.5]
cap_upperq = cap_quartiles[0.75]
cap_iqr = cap_upperq-cap_lowerq
cap_lowerbound = cap_lowerq - (1.5*cap_iqr)
cap_upperbound = cap_upperq + (1.5*cap_iqr)

# Ramicane quartiles and IQR
ram_quartiles = Ramicane_tumorVol.quantile([.25,.5,.75])
ram_lowerq = ram_quartiles[0.25]
ram_median = ram_quartiles[0.5]
ram_upperq = ram_quartiles[0.75]
ram_iqr = ram_upperq-ram_lowerq
ram_lowerbound = ram_lowerq - (1.5*ram_iqr)
ram_upperbound = ram_upperq + (1.5*ram_iqr)

# Infubinol quartiles and IQR
inf_quartiles = Infubinol_tumorVol.quantile([.25,.5,.75])
inf_lowerq = inf_quartiles[0.25]
inf_median = inf_quartiles[0.5]
inf_upperq = inf_quartiles[0.75]
inf_iqr = inf_upperq-inf_lowerq
inf_lowerbound = inf_lowerq - (1.5*inf_iqr)
inf_upperbound = inf_upperq + (1.5*inf_iqr)

# Ceftamin quartiles and IQR
cef_quartiles = Ceftamin_tumorVol.quantile([.25,.5,.75])
cef_lowerq = cef_quartiles[0.25]
cef_median = cef_quartiles[0.5]
cef_upperq = cef_quartiles[0.75]
cef_iqr = cef_upperq-cef_lowerq
cef_lowerbound = cef_lowerq - (1.5*cef_iqr)
cef_upperbound = cef_upperq + (1.5*cef_iqr)
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Creating the dataframe with mouse weight and tumor volume
capomulinmouse_df = mousestudy_cleaned[mousestudy_cleaned["Drug Regimen"]=='Capomulin']
capomulinmousereduced_df = capomulinmouse_df[['Mouse ID', 'Tumor Volume (mm3)', 'Weight (g)']]
cap_grpd = capomulinmousereduced_df.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean().reset_index()

cap_grpd_df = pd.DataFrame(cap_grpd)
cap_grpd_df
```

```python
#Plotting the scatter plot
cap_grpd_df.plot(kind="scatter", x='Weight (g)', y="Tumor Volume (mm3)", grid=True, figsize=(8,8),
              title="Average tumor volume vs. mouse weight for the Capomulin regimen")
plt.xlim(10, 30)
plt.ylim(30, 50)
plt.show()
```

```python
#Calculate the correlation coefficient and linear regression model between mouse weight and average tumor volume for the Capomulin treatment. 
#Plot the linear regression model on top of the previous scatter plot.

# Calculating Correlation coefficient:
mouse_wgt = cap_grpd_df['Weight (g)']
avg_tumvol = cap_grpd_df['Tumor Volume (mm3)']
correlation = round(sts.pearsonr(mouse_wgt, avg_tumvol)[0],2)
print(f"The correlation between mouse weight and average tumor volume for the Capomulin treatment is {correlation}")
```

```python
# Calculating the slope, y-intercept to build the regression model.

x_val = cap_grpd_df['Weight (g)']
y_val = cap_grpd_df['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_wgt, avg_tumvol)
regress_val = mouse_wgt * slope + intercept

regLineEq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"Slope: {slope}, Intercept: {intercept}")
print(f"{regLineEq}")
```

```python
# Plotting the regression line
plt.scatter(mouse_wgt, avg_tumvol)
plt.plot(mouse_wgt,regress_val,"r-")
plt.annotate(regLineEq,(6,10), fontsize=15, color="g",)
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.show()
```
